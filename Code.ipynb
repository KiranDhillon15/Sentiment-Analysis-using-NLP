```python
data['cleaned_review'] = data['review'].apply(lambda text: re.sub('<.*?>', '', text))  # Remove HTML tags
data['cleaned_review'] = data['cleaned_review'].apply(lambda text: re.sub('[^a-zA-Z0-9\s]', '', text))  # Remove non-alphanumeric characters
data['cleaned_review'] = data['cleaned_review'].apply(lambda text: re.sub('\s+', ' ', text).strip())  # Remove extra whitespace
data['cleaned_review'] = data['cleaned_review'].apply(lambda text: text.lower())  # Convert to lowercase
for review in data['cleaned_review'].head(10):
    print(review)
    print('---')
```

```python
file_path = '/content/IMDB Dataset.csv'
data = pd.read_csv('IMDB Dataset.csv')
reviews = data['review'].tolist()
labels = data['sentiment'].tolist()

# Tokenize the dataset
for review in data['review'].head(100):
    tokens = word_tokenize(review)
    print(tokens)
    print('---') #tokenized output
```

```python
# Remove stop words from the dataset
stop_words = set(stopwords.words('english'))

for review in data['review'].head(100):
    tokens = word_tokenize(review)
    filtered_tokens = [token for token in tokens if token.lower() not in stop_words]
    print(filtered_tokens)
    print('---') # Now the output of stop words removal for 100 reviews.
```

```python
#pos tags
for review in data['review'].head(10):
    tokens = word_tokenize(review)
    pos_tags = pos_tag(tokens)
    print(pos_tags)
    print('---')
```

```python
#lemmatization
lemmatizer = WordNetLemmatizer()

for review in data['review'].head(10):
    tokens = word_tokenize(review)
    lemmas = [lemmatizer.lemmatize(token) for token in tokens]
    print(lemmas)
    print('---')
```

```python
#Spliting the dataset
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(reviews, labels, test_size=0.3, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
from gensim.models import Word2Vec
data_subset = data.head(100)
# Bag of Words (BoW) feature extraction
bow_vectorizer = CountVectorizer()
bow_features = bow_vectorizer.fit_transform(data_subset['review'])

# Print the output
print("Bag of Words (BoW) features:")
print(bow_features.toarray())
```

```python
# TF-IDF feature extraction
tfidf_vectorizer = TfidfVectorizer()
tfidf_features = tfidf_vectorizer.fit_transform(data_subset['review'])
print("TF-IDF features:")
print(tfidf_features.toarray())
```

```python
# Word Embeddings (Word2Vec) feature extraction
sentences = [review.split() for review in data_subset['review']]
word2vec_model = Word2Vec(sentences, min_count=1)
word2vec_features = [word2vec_model.wv[words] for words in sentences]
print("Word Embeddings (Word2Vec) features:")
for feature in word2vec_features:
    print(feature)
```

```python
#Training
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding, LSTM, Dense
from tensorflow.keras.optimizers import Adam
import tensorflow as tf
```

```python
# Naive Bayes
from sklearn.naive_bayes import MultinomialNB
nb_vectorizer = CountVectorizer()
X_train_nb = nb_vectorizer.fit_transform(X_train)
X_test_nb = nb_vectorizer.transform(X_test)
nb_model = MultinomialNB()
nb_model.fit(X_train_nb, y_train)
nb_predictions = nb_model.predict(X_test_nb)
nb_accuracy = accuracy_score(y_test, nb_predictions)
print("Naive Bayes Accuracy:", nb_accuracy)
```

```python
# Recurrent Neural Networks (RNN)
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_train)
X_train_rnn = tokenizer.texts_to_sequences(X_train)
X_test_rnn = tokenizer.texts_to_sequences(X_test)
vocab_size = len(tokenizer.word_index) + 1
max_length = max(len(sequence) for sequence in X_train_rnn)
X_train_rnn = pad_sequences(X_train_rnn, maxlen=max_length, padding='post')
X_test_rnn = pad_sequences(X_test_rnn, maxlen=max_length, padding='post')
le = LabelEncoder()
y_train_rnn = le.fit_transform(y_train)
y_test_rnn = le.transform(y_test)
rnn_model = Sequential()
rnn_model.add(Embedding(vocab_size, 100, input_length=max_length))
rnn_model.add(LSTM(100))
rnn_model.add(Dense(1, activation='sigmoid'))
rnn_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
rnn_model.fit(X_train_rnn, y_train_rnn, validation_data=(X_test_rnn, y_test_rnn), epochs=5)
_, rnn_accuracy = rnn_model.evaluate(X_test_rnn, y_test_rnn)
print("RNN Accuracy:", rnn_accuracy)
```
